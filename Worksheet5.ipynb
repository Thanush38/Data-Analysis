```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
d_diam = sns.load_dataset('diamonds')
d_ice = sns.load_dataset('seaice')

sns.scatterplot(data=d_diam, x='carat', y='price', hue="cut").set( xlabel='Weight (carats)', ylabel='Price ($)', title='The Price of Diamonds')

```

```python
d_ice.Extent.plot(kind='line', linestyle='dashed', color='black')
xpos = [0,3000,6000, 9000,12000]
label = [1980, 1992,2000,2008,2016]
plt.xticks(xpos,label)
plt.set('Change in Artic sea ice thickness over time')
plt.xlabel('Year')
plt.ylabel('Thickness(m)')
```
