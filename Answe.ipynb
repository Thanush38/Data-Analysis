```python
import seaborn as sns
import matplotlib.pyplot as plt

d_diam = sns.load_dataset('diamonds')
d_ice = sns.load_dataset('seaice')

sns.scatterplot(x = 'carat', y = 'price', hue = 'cut', data = d_diam)

plt.xlabel("Weight(carats)")
plt.ylabel("Price($)")
plt.title("The price of diamonds")
```

```python
d_ice.Extent.plot(kind = 'line', linestyle = 'dashed', color = 'black')

xtick_positions = [0, 3000, 6000, 9000, 12000]
xtick_labels = ["1980", "1992", "2000", "2008", "2016"]
plt.xticks(xtick_positions,xtick_labels)

plt.xlabel("Year")
plt.ylabel("Thickness(m)")
plt.title("Change in arctic sea ice thickness over time")
```
